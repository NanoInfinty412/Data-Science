# Project 11: Fourier Series 


## Abstract & description:

Fourier series is mathematical technique allowing to represent periodic function with the sum of cosines and sines.
In our project, we are going to investigate a square wave and try to represent it with a finite sum of sines and cosines. The goal is to have a graph that is plus or minus matching our initial square wave.
IN order to plot using this technique, we have to integrate. 
In our case, the integral that we solved led to the following mathematical expression:

  $$ f = \sum_{n=1}^{\infty}\frac{-2\alpha}{n\pi}\sin( \frac{n\pi t}{2T}) $$


## Algorithm and Discussion

In this experiment, we are going to consider a simple square wave function and try with different techniques to obtain a specific periodic square wave.


Suppose f is a periodic function with a period T = 2L. Then the Fourier
series representation of f is a trigonometric series (that is, it is an infinite
series consists of sine and cosine terms of the form.

Basically, here is how the equations are presented for a simple square wave.



$$f(x) = \frac{a_0}{2}+\sum_{i=1}^{\infty}(a_n\cos \frac{(n\pi x)}{L}+b_n\sin \frac{(n\pi x)}{L})$$

The coefficients are given by:

$$a_n = \frac{1}{L}\int_{-L}^{L} f(x)\cos \frac{(m\pi x)}{L}dx, \space m =0,1,2,...$$


$$b_n = \frac{1}{L}\int_{-L}^{L} f(x)\sin \frac{(n\pi x)}{L}dx, \space n =0,1,2,...$$

## Conclusion

Altogether, we discovered how Fourier series work. It starts as simple sines or cosines and as we keep adding other factors, the graph keeps wiggling. In our case, we wanted a square wave, which is composed of straight lines. We ended up obtaining straight figures by applying the idea the following idea:
If we have a large number of wiggles of the same amplitude  in a very short path, they all line up and produce a straight line.

```python
y = np.sin(t) + np.sin(3*t)/3 + np.sin(5*t)/5 + np.sin(7*t)/7 + np.sin(9*t)/9;
plt.plot(t,y);
```

As we keep adding odd harmonics, we will eventually reach a big number and the top of our square wave will be a straight line.
