```python
import pandas as pd
```

```python
df=pd.read_csv("Pend5degrees.csv")
df.head()
```

```python
df.plot("Time","Angle")
```

```python
times = df[df['Time']>19].Time.values
angles = df[df['Time']>19].Angle.values
plot(times, angles, 'r+')
```

```python
from scipy.optimize import curve_fit
```

```python
def sinFit(t, A, omega, phi, y0 ):
    return A*sin(omega*t+phi)+y0

A0=6            # amplitude is about ~6 degrees
y00=0           # y0 is about zero
T0=7.25/4       # period is roughly 7.25 sec/4
omega0=2*pi/T0  # omega is 2*pi/T
phi0=-omega0*19.25 # phase is whatever it taks to have sin go through zero at 19.25 sec.

params, cov = curve_fit(sinFit, times, angles, p0=(A0, omega0, phi0, y00), sigma=360.0/2048)

A,omega,phi,y0=params
print(params)

tp=linspace(times[0],times[-1],100)
yp=sinFit(tp, A, omega, phi, y0)
errorbar(times, angles, 0.2, fmt='r.')
plot(tp, yp, 'b-')

```
